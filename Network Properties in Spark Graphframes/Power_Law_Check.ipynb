```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import os
import pandas as pd
%matplotlib inline
```

```python
Root_direc = '/home/vivekreddy/GDM_P1/results/'
list_files = []
for subdir, dirs, files in os.walk(Root_direc):
    for file in files:
        list_files.append(file)
print(files)
```

```python
from sklearn.linear_model import LinearRegression

def check_power_law(path_to_csv):
    data = pd.read_csv(path_to_csv, header=0).to_numpy()
    data = np.log10(data)
    LR = LinearRegression().fit(data[:,1].reshape(-1, 1) , data[:,2].reshape(-1, 1))
    print("For the graph {}".format(path_to_csv.split("/")[-1]))
    print("Coefficient is {}".format(str(LR.coef_)))
    print("Intercept is {}".format(str(LR.intercept_)))
    print("R^2 Coeffecient is {}".format(LR.score(data[:,1].reshape(-1, 1) , data[:,2].reshape(-1, 1))))
    plt.title("The Log-log plot for Graph: {}".format(path_to_csv.split("/")[-1]))
    plt.plot(data[:,1], data[:,1]*LR.coef_[0][0]+LR.intercept_[0])
    plt.plot(data[:,1], data[:,2])
    plt.show()
    
    return LR
```

```python
for file in files:
    check_power_law(Root_direc+file)
```

```python
def display_data(path_to_csv):
    data = pd.read_csv(path_to_csv, header=0).to_numpy()
    plt.title("The Graph: {}".format(path_to_csv.split("/")[-1]))
    plt.plot(data[:,1], data[:,2])
    plt.show()  
    return None
for file in files:
    display_data(Root_direc+file)
```
